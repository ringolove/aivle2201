## 2.대푯값

1) 평균

```python
# 산술평균
a = [1, 2, 3, 3, 4, 4, 4, 5, 6, 7, 8, 9, 10, 11, 15, 17, 17, 19, 20, 25, 30]
print(np.mean(a))

# 조화평균
# 거리 20 km
# 갈때 속력 60 km
# 올때 속력 80 km
# 평균 속력은?

a1, a2 = 60, 80
hm = 2*a1*a2 / (a1 + a2)
print(hm)
```

2) 중앙값(중위수, median)  

자료의 순서상 가운데 위치한 값

```python
a = [1, 2, 3, 3, 4, 4, 4, 5, 6, 7, 8, 9, 10, 11, 15, 17, 17, 19, 20, 25, 30]
print(np.median(a))
```

3) 최빈값

자료 중에서 가장 빈번하게 나타나는 값(빈도가 높은 값)

```python
a = [1, 2, 3, 3, 4, 4, 4, 5, 6, 7, 8, 9, 10, 11, 15, 17, 17, 19, 20, 25, 30]
a = pd.Series(a)
a.mode()
```

4) 4분위수

![](https://mblogthumb-phinf.pstatic.net/20150702_229/jevida_1435803723327eyeeG_PNG/3.png?type=w2)

```python
a = [1, 2, 3, 3, 4, 4, 4, 5, 6, 7, 8, 9, 10, 11, 15, 17, 17, 19, 20, 25, 30]
print(np.percentile(a, [0, 25, 50, 75, 100]))
```

연습문제를 풀어 봅시다.

```python
l1 = list(range(0, 20,5)) + list(range(0,20,2)) + [10,11,12,15,15,17]
# l1의 평균
print(np.mean(l1))

# l1의 중앙값
print(np.median(l1))

# l1의 최빈값
print(pd.Series(l1).mode())

# l1의 사분위수
print(np.percentile(l1, [0, 25, 50, 75, 100]))
```
