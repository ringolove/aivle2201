6) [심화]반복문을 이용하여 100이하의 소수(prime number)를 리스트에 담아 봅시다.
* 소수(prime number) : 약수가 오직 1과 자기자신 만 존재하는 수. 

```python
def prime(i):
    global prime
    prime = [False,False] + [True]*(i-1)
    for i in range(2,i+1):
        if prime[i]:
            for j in range(2*i,i+1,i):
                prime[j]

def prima(i):
    global prime
    return prime[i]

n=100
a=[]
prime(n)
for i in range(n):
    if(prima(i)):
        a.append(i)
print(a)

```

7) [심화] 6번의 코드를 참조하여, 소수(prime number) 추출기 함수를 만들어 봅시다.
* 입력변수 : 숫자 n
* 처리 : n 이하의 소수를 리스트에 저장
* 출력변수 : 소수 리스트

* [추가] 입력변수 n은 2이상의 정수여야 함.(입력값 체크 부분 포함)

```python
def prime(i):
    global prime
    prime = [False,False] + [True]*(i-1)
    for i in range(2,i+1):
        if prime[i]:
            for j in range(2*i,i+1,i):
                prime[j]

def prima(i):
    global prime
    return prime[i]

n=int(input("2 이상의 숫자 입력: "))
a=[]
if n > 1:
    prime(n)
    for i in range(n):
        if(prima(i)):
            a.append(i)
else:
    print("1보다 큰 정수를 입력: ")
    
print(a)
```
