```python
import pandas as pd
import requests
```

```python
from datetime import datetime

def krflix_rank(date):
  url = f'https://top10.netflix.com/_next/data/5HatR7zO1WshaXilzP8tj/south-korea/films/{date}.json'
  response = requests.get(url)
  data = response.json()['pageProps']['data']['weeklyTopTen']
  return pd.DataFrame(data).iloc[:, -3].rename(date)

# 2021년 7월 4일부터 조회 가능
movies = []
date = pd.date_range(datetime(year=2021,month=7,day=4), datetime(year=2021,month=12,day=26), freq='W')
date_str = [date_obj.strftime('%Y-%m-%d') for date_obj in date]

for date in date_str:
  movies.append(krflix_rank(date))

pd.concat(movies, axis=1)
```

```python
df = pd.DataFrame(movies)
df.head(5)
```

```python
df_reversed = df.transpose()
df_reversed = df.T
df_reversed
```
